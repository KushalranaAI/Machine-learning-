Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and Data Preprocessing

```python
# loading the dataset to pandas data frame
sonar_data = pd.read_csv('/content/sonar data.csv', header = None)
```

```python
sonar_data.head()
```

```python
# number of rows and columns
sonar_data.shape
```

```python
sonar_data.describe()  # describe --> give statistical measures of data
```

```python
sonar_data[60].value_counts()
```

M --> mine

R --> Rock

```python
sonar_data.groupby(60).mean()
```

```python
# separate the data and label
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60] 
```

```python
print(X)
```

Training and testing data spliting

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1, stratify =Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model trainig --> logistic Regression

```python
model = LogisticRegression()
```

```python
# training the logical regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
# accuracy on the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data: ",training_data_accuracy)
```

```python
# accuracy on the test data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on testing data: ", testing_data_accuracy)
```
