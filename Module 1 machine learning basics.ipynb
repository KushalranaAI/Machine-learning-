# What is a AI?
Artificial intelligence is a branch of computer science that is concerned with building smart & intelligent Machince
- what is intelligent machine? An intelligent machine is a machine that can perform tasks that typically require human intelligence, such as understanding natural language, recognizing images, or making decisions.
- what is non-intelligent machine? A non-intelligent machine, on the other hand, is a machine that can perform specific, limited tasks, but does not have the ability to think or learn like a human. For example, a washing machine is a non-intelligent machine because it can only wash clothes, but it does not have the ability to understand language or make decisions.

# Types of machine learning
There are three main types:
- Supervised Learning
- Usupervised learning
- Reinforcement learning
## Supervised learning
In supervised learning the machine learning algorithm learns from labelled data.
- Labelled data means Labeled data is a set of data that has been labeled(tags), or annotated, with the correct output or outcome. The data is used in supervised machine learning, where the goal is to train a model that can make predictions on new, unseen data. The labeled data provides the model with information about the relationship between inputs (e.g. features or attributes) and outputs (e.g. labels or classifications), allowing the model to learn how to make predictions

- The labeled data is usually split into two sets: a training set, which is used to train the model, and a testing set, which is used to evaluate the performance of the trained model.





## Unsupervised learning
In unsupervised learning, the machine learning algorithm learn from unlabelled data. we give input to algorithm without telling what it is and what it can do?



## Reinforcement learning
Reinforcement learning is an area of machine learning concerned with how intelligent agents take actions in an environment to maximize it's rewards.
There are 3 main aspects :
- Environment
- Agent
- Action
- Rewards
- The agent's goal is to learn a policy that maximizes the total reward over time. This type of learning is often used in control systems, gaming and decision making scenarios where the agent learns to optimize its behavior through trial and error.

# Machine learning vs Deep learning

The main difference between these two are Feature extraction.Feature extraction means that is a technique used to extract relevant information or features from raw data in order to represent it in a more meaningful and compact form. The extracted features can then be used for tasks such as classification, clustering, and regression. Feature extraction can be done manually, by an expert in the domain, or automatically, using techniques such as dimensionality reduction, feature selection, and feature engineering. Commonly used techniques include Principal Component Analysis (PCA), Linear Discriminant Analysis (LDA), and Independent Component Analysis (ICA).

Deep learning doesn't need to dd feature extraction manually.
